```python
import networkx as nx
import re
import string
```

```python
# Create an empty graph
G = nx.Graph()

# Split the sentence into words
sentence = "This is an example sentence."
words = sentence.split()

# Add edges between words
for i in range(len(words) - 1):
    G.add_edge(words[i], words[i+1])

# Draw the graph
nx.draw(G, with_labels=True)
```

```python
import networkx as nx
import re

# Create an empty graph
G = nx.Graph()

# Define the sentence
sentence = "Artificial intelligence is the simulation of human intelligence processes by machines."

# Use regular expressions to find all words in the sentence
words = re.findall(r'\b\w+\b', sentence)


# Add nodes to the graph
for word in words:
    G.add_node(word)

# Add edges between words
for i in range(len(words) - 1):
    G.add_edge(words[i], words[i+1])

# Draw the graph
nx.draw(G, with_labels=True)
```

```python
# Create an empty graph
G = nx.Graph()
# Define the sentence
sentence = "Artificial intelligence is the simulation of human intelligence processes by machines."
#cleaning text without stemming
sentence = sentence.lower()
sentence = sentence.split(" ")


# Add nodes to the graph
for word in sentence:
    G.add_node(word)

# Add edges between words
for i in range(len(sentence) - 1):
    G.add_edge(sentence[i], sentence[i+1])

# Draw the graph
nx.draw(G, with_labels=True)
```
