# **Exploring the data**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

# **Models and evaluation**

**Kmeans (partitioning)**

```python
from sklearn.cluster import KMeans
X = data2
distortions = []
K = range(1,5)
for k in K:
  kmeans = KMeans(n_clusters=k)
  kmeans.fit(X)
  distortions.append(kmeans.inertia_)

plt.figure(figsize=(10,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()

kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
```

```python
from sklearn import metrics
print(metrics.davies_bouldin_score(X,kmeans.labels_))
print(metrics.calinski_harabasz_score(X,kmeans.labels_))
print(metrics.silhouette_score(X,kmeans.labels_))


dbs = metrics.davies_bouldin_score(X,kmeans.labels_)
chs = metrics.calinski_harabasz_score(X,kmeans.labels_)
ss = metrics.silhouette_score(X,kmeans.labels_)
models = {'kmeans':{'davies_bouldin_score':dbs ,'calinski_harabasz_score': chs, 'silhouette_score':ss}}
data = pd.DataFrame(data=models)
data.head()
```

**DBSCAN (density based)**

```python
from sklearn.cluster import DBSCAN
clustering = DBSCAN(eps=4, min_samples=2)
clustering.fit(X)
print(clustering.labels_)
```

```python
print(metrics.davies_bouldin_score(X,clustering.labels_))
print(metrics.calinski_harabasz_score(X,clustering.labels_))
print(metrics.silhouette_score(X,clustering.labels_))

dbs = metrics.davies_bouldin_score(X,clustering.labels_)
chs = metrics.calinski_harabasz_score(X,clustering.labels_)
ss = metrics.silhouette_score(X,clustering.labels_)
models2 = [chs , dbs, ss]
data['dbscan'] = models2
data
```

```python
for x in range(data2.shape[1]):
  fig = plt.figure()
  ax = fig.add_subplot(111)
  scatter = ax.scatter(data2.iloc[:,x],data2.iloc[:,x+1],c=clustering.labels_,s=50)
  ax.set_title('DBSCAN')
  ax.set_xlabel(data2.columns[x])
  ax.set_ylabel(data2.columns[x+1])
  if(x==6):
    break
  plt.colorbar(scatter)
```
